# Twitter Data Compilation

This notebook is used to combine all the twitter data from various excel files into a single dataframe. Then to prepare data and save into a new file for the analysis.

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

### Compiling Excel data together

```python
base_path = '../data/excel_files/'
file_1 = 'NodeXLGraph1TwitterUBuffalo (2).xlsx'
file_2 = 'Twitter data for hashtags (2).xlsx'
file_3 = 'UBAlumni_3.13.22.xlsx'
file_4 = 'UBuffaloTwitter3-14-22.xlsx'
```

```python
df_1 = pd.read_excel(base_path+file_1, skiprows=1)
df_2 = pd.read_excel(base_path+file_2)
df_3 = pd.read_excel(base_path+file_3, skiprows=1)
df_4 = pd.read_excel(base_path+file_4, skiprows=1)
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
df_3.head()
```

```python
df_4.head()
```

```python
df_1.columns
```

```python
df_1_imp = df_1[['Unified Twitter ID', 'Tweet','Hashtags in Tweet', 'Language']]
df_1_imp.drop_duplicates(subset='Unified Twitter ID', keep='first', inplace=True, ignore_index=True)
df_1_imp.head()
```

```python
df_2.columns
```

```python
df_2_imp = df_2[['Hashtag','Text']]
df_2_imp.drop_duplicates(subset='Text', keep='first', inplace=True, ignore_index=True)
df_2_imp.head()
```

```python
df_3.columns
```

```python
df_4_imp = df_4[['Unified Twitter ID', 'Tweet','Hashtags in Tweet', 'Language']]
df_4_imp.drop_duplicates(subset='Unified Twitter ID', keep='first', inplace=True, ignore_index=True)
df_4_imp.head()
```

Combining DF 1 and DF 4 and then extracting only english language rows

```python
df_comb = pd.concat([df_1_imp,
                   df_4_imp])
```

```python
df_comb.drop_duplicates(subset='Unified Twitter ID', keep='first', ignore_index=True, inplace=True)
df_comb.drop(columns=['Unified Twitter ID'], inplace=True)
df_comb.head()
```

```python
df_eng = df_comb[df_comb['Language']=='en']
df_eng['Language'].unique()
```

```python
df_comb.drop(columns=['Language'], inplace=True)
df_comb.head()
```

```python
df_2_imp.rename(columns={'Text': 'Tweet'}, inplace=True)
df_comb.rename(columns={'Hashtags in Tweet': 'Hashtag'}, inplace=True)
```

```python
df_2_imp = df_2_imp[df_comb.columns.tolist()]
df_2_imp.head()
```

```python
df_comb.head()
```

```python
df_final = pd.concat([df_comb,
                     df_2_imp])
```

```python
df_final.shape
```

```python
len(df_final.Tweet.unique())
```

```python
df_final.drop_duplicates(subset='Tweet', keep='first', ignore_index=True, inplace=True)
df_final.shape
```

```python
df_final.head()
```

```python
hashtags_df_final = []

for data in df_final['Hashtag'].dropna():
    for word in data.split(' '):
        hashtags_df_final.append(word.strip("'#@*&,:;/"))
```

```python
len(hashtags_df_final)
```

```python
new_file_name = 'combined_twitter_df.csv'
```

```python
df_final.to_csv(base_path+new_file_name)
```
